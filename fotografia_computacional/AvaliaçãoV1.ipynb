```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("imgcourse/lena.png")
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img_gray, cmap='gray')
plt.show()
```

```python
img_fft = np.fft.fft2(img_gray)

img_fshift = np.fft.fftshift(img_fft)
img_spectrum = np.abs(img_fshift)
img_spectrum_log = np.log(1 + img_spectrum)
```

```python
plt.subplot(1, 2, 1), plt.imshow(img_spectrum, cmap='gray'), plt.title('Spectrum')
plt.subplot(1, 2, 2), plt.imshow(img_spectrum_log, cmap='gray'), plt.title('Spectrum log')
plt.show()
```

```python
print(np.mean(img_spectrum_log))
```

```python
img_equalized = cv2.equalizeHist(img_gray)

plt.subplot(1,2,1), plt.imshow(img_gray, cmap='gray'), plt.title('Original image')
plt.subplot(1,2,2), plt.imshow(img_equalized, cmap='gray'), plt.title('Equalized image')
plt.show()
```

```python
hist_equalized = cv2.calcHist([img_equalized],[0],None,[256],[0,256])
hist_original = cv2.calcHist([img_gray],[0],None,[256],[0,256])

plt.subplot(1,2,1), plt.plot(hist_original), plt.title('Original histogram')
plt.subplot(1,2,2), plt.plot(hist_equalized), plt.title('Equalized histogram')
plt.show()
```

```python
print(np.mean(img_equalized))
print(np.std(img_equalized))
```
